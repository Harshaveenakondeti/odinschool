```python
# sum of n numbers
n=int(input('enter num: '))
sum=0
i=1
while i<=n:   # when i==n it will stop
  sum= sum+i  # sum till n numbers
  print(sum,end=" ")
  i=i+1
print(f'\n sum of {n} numbers is {sum}')
```
