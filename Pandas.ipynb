## Accesing Dataframes
you can access the rows of data frames using location(loc) and index location(iloc), columns u can access directly

```python
df=pd.DataFrame(np.random.randint(10,100,(10,4)),columns = list("ABCD"))  # create datframe (start,end,(table rows,columns))
df
```

```python
df['A'] # columns access direcltly
```

```python
 # rows can be accessed with loc or row location fun
 df.loc[::]     # both start func,end posion will be included in loc func
```

```python
df.loc[:4:]
```

```python
df.loc[:4:2]  # row index4.step index 2
```

```python
df.loc[[1,3,5],:]  #only give wanted rows
```

```python
df.loc[:,['A','C']]  # with particular column names
```

```python
df.loc[[1,3,5,7],['A','D']] # with particular rows n columns
```

```python
df.loc[1:6,'A':'C']  # from 1 to 6 rows and A o C columns
```

```python
# dates
dates= pd.date_range(start="2022-12-4",periods=10, freq="D")
dates
```

```python
df1=pd.DataFrame(np.random.randint(10,100,(10,4)),columns = list("ABCD"),index=dates)  # create datframe (start,end,(table rows,columns))
df1
```

```python
df1.loc["2022-12-04":"2022-12-08",:]   # loc can access name of index ,not position of index
```

```python
df1.iloc[2:5,1:] # iloc means index location will exclude end position
```

```python
df1.iloc[[1,3,5],[0,2]]  # this will give particular number of rows n columns
```

```python
df1.iloc[1:4,2:]   # accessing 1 to 4 and 2nd column to all
```

## Boolean Indexing
when we lookout for particular info,
we have to select subset of data based on the actual values of data in dataframe and not based on their row or column. labels or integer locations

```python
students['Age']>10  # gives boolean output of particular condition
```

```python
# to get values satisfyig condition
students[students['Age']>13]
```

```python
students[students['Age']>13]["names"]   # only gives names satisfying condiion
```

```python
students[students['Age']>13][["names","marks"]] 
```

```python
students["marks"].mean()  # mean of marks
```

```python
aboveavg= students["marks"] > students["marks"].mean()
```

creating aboveavg variable, condition values will assign in to it

```python
students[aboveavg]
```

```python
students[aboveavg].sort_values('marks')
```
